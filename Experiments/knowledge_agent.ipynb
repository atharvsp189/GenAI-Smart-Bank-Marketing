```python
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://www.paisabazaar.com/sbi-bank/"

# Fetch the page
response = requests.get(url)
response.raise_for_status()  # raise error if request failed

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Extract all <a> tags with href
links = [a["href"] for a in soup.find_all("a", href=True)]

```

```python
for link in links:
    print(link)
```

```python
import requests

def is_reachable_url(url: str) -> bool:
    try:
        response = requests.head(url, allow_redirects=True, timeout=5)
        return response.status_code < 400  # 2xx or 3xx considered valid
    except requests.RequestException:
        return False

for link in links:
    print("Is Link Reachable: ", is_reachable_url(link))
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(["https://www.example.com/", "https://google.com"])
loader.requests_per_second = 1
docs = loader.aload()
docs
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["google_api_key"] = os.getenv("GOOGLE_API_KEY")
os.environ["qdrant_api_key"] = os.getenv("Qdrant_API_KEY")
qdrant_url = os.getenv("Qdrant_URL")
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_google_genai import GoogleGenerativeAIEmbeddings

# Initialize the embedding model
embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")

# Initialize the SemanticChunker
text_splitter = SemanticChunker(embeddings)

# Create documents (chunks) from a large text
chunks = text_splitter.create_documents([docs[1].page_content])

# Now 'docs' contains a list of semantically coherent document chunks
for chunk in chunks:
    print(chunk.page_content)
    print("-" * 20)
```
